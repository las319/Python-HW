```python
#Combine/merge patients df with observations df
#need to rename PATIENT in observations to Id to match df_patients column header

df_observations.columns = ["DATE","Id","ENCOUNTER","CODE", "DESCRIPTION","VALUE", "UNITS", "TYPE"]
df_observations.columns
```

```python
#confirm that column was renamed as "Id"
df_observations.head()
```

```python
#merge two dataframes
df_merged = pd.merge(df_patients, df_observations, on="Id")
df_merged.head()
```

```python
#action 1: reset the index to be the Id
pd.DataFrame(df_merged).set_index(pd.DataFrame(df_merged)['Id'])
```

```python
#action 2: find the oldest patient based on birthdate year
df_merged["BIRTHDATE"].min()
```

```python
#action 3: identify how many unique Ids (patients) there are in set
len(df_merged["Id"].unique())
```

```python
#action 4: subset by QOL 
query_QOL = df_merged["DESCRIPTION"] == "QOL" 
query_QOL.sample(50)
```

```python
#action 5: make new dataframe from subset
df_QOL = df_merged[query_QOL]
df_QOL.head()
```

```python
#Action 6: see the counts for different QOL scores to understand spread within this patient sample
df_QOL["VALUE"].value_counts()
```
