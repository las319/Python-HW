```python
#import csv and Pandas
import csv
import pandas as pd
```

```python
#create a DataFrame using Pandas from the csv file
df = pd.read_csv("ex_diabetes_rural.csv")
#read/see the data to gain understanding of what the data is looking at
df
```

```python
#exploratory analysis goal: compare average final FPG between family history groups to identify if family history is associated with higher FPG levels/T2DM 
#action 1: group by family history using .groupby function
df_hist = df.groupby("family histroy of diabetes(1,Yes;0,No)")
df_hist.sum()
```

```python
#action 2: transform FPG from mmol/L to mg/dL - need to make a new column by vectorized math
df["FPG of final visit(mg/dL)"] = df["FPG of final visit(mmol/L)"] * 18
df_hist. head()
```

```python
#action 3: find average FPG for reference
df["FPG of final visit(mg/dL)"].mean()
```

```python
#action 4: compare average FPG between family history groups
df_hist['FPG of final visit(mg/dL)'].mean()
```

```python
#action 5: what is the maximum and minimum FPG between family history groups
df_hist['FPG of final visit(mg/dL)'].max()
```

```python
#action 5: what is the maximum and minimum FPG between family history groups
df_hist['FPG of final visit(mg/dL)'].min() 
```

```python
#action 6: see sample size for each family history group to see if comparison is valid
df_hist.count()
```
